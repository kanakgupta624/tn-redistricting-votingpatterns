```python
import pandas as pd
```

2018

```python
df18 = pd.read_csv('../data/Nov 2018 General results.csv')
df18
```

```python
gov18 = df18[df18['OFFICENAME'] == 'Governor'].copy()
gov18.shape
```

```python
vote_cols_18 = [c for c in gov18.columns if c.startswith('PVTALLY')]
```

```python
gov18['total_votes'] = gov18[vote_cols_18].sum(axis=1)
```

```python
gov18[['COUNTY', 'PRECINCT', 'total_votes']].head()
```

```python
def clean_text(col):
    return col.str.upper().str.strip()
```

```python
for df in [gov18]:
    df['COUNTY'] = clean_text(df['COUNTY'])
    df['PRECINCT'] = clean_text(df['PRECINCT'])
```

```python
county18 = gov18.groupby('COUNTY', as_index=False)['total_votes'].sum()
county18
```

```python
county18.to_csv('cleaned_2018_governor_turnout_by_county.csv', index=False)
```
