Governor Votes by Party and County

```python
party_vote_pairs = [(f"PARTY{i}", f"PVTALLY{i}") for i in range(1, 11)]
party_vote_pairs
```

```python
def reshape_party_votes(df, year):
    rows = []
    
    for party_col, vote_col in party_vote_pairs:
        subset = df[["COUNTY", party_col, vote_col]].copy()
        subset = subset.dropna(subset=[vote_col])
        subset = subset.rename(columns={party_col: "party", vote_col: "votes"})
        subset["year"] = year
        rows.append(subset)

    return pd.concat(rows, ignore_index=True)
```

```python
party_2018 = reshape_party_votes(gov_2018, 2018)
party_2018
```

```python
party_2022 = reshape_party_votes(gov_2022, 2022)
party_2022
```

```python
for df in [party_2018, party_2022]:
    df["party"] = df["party"].str.upper().str.strip()
```

```python
county_party_2018 = (party_2018.groupby(["COUNTY", "party", "year"], as_index=False)["votes"].sum())
county_party_2018
```

```python
county_party_2022 = (party_2022.groupby(["COUNTY", "party", "year"], as_index=False)["votes"].sum())
county_party_2022
```

```python
county_totals = (county_party_all.groupby(["COUNTY", "year"], as_index=False)["votes"].sum().rename(columns={"votes": "total_votes"}))
```
