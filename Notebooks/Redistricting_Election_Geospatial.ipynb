Governor Votes by Party and County

```python
party_vote_pairs = [(f"PARTY{i}", f"PVTALLY{i}") for i in range(1, 11)]
party_vote_pairs
```

```python
def reshape_party_votes(df, year):
    rows = []
    
    for party_col, vote_col in party_vote_pairs:
        subset = df[["COUNTY", party_col, vote_col]].copy()
        subset = subset.dropna(subset=[vote_col])
        subset = subset.rename(columns={party_col: "party", vote_col: "votes"})
        subset["year"] = year
        rows.append(subset)

    return pd.concat(rows, ignore_index=True)
```

```python
party_2018 = reshape_party_votes(gov_2018, 2018)
party_2018
```

```python
party_2022 = reshape_party_votes(gov_2022, 2022)
party_2022
```

```python
for df in [party_2018, party_2022]:
    df["party"] = df["party"].str.upper().str.strip()
```

```python
county_party_2018 = (party_2018.groupby(["COUNTY", "party", "year"], as_index=False)["votes"].sum())
county_party_2018
```

```python
county_party_2022 = (party_2022.groupby(["COUNTY", "party", "year"], as_index=False)["votes"].sum())
county_party_2022
```

```python
county_party_2018["party"].value_counts()
```

```python
county_party_2022["party"].value_counts()
```

```python
county_party_all = pd.concat([county_party_2018, county_party_2022],ignore_index=True)
county_party_all
```

```python
county_totals = (county_party_all.groupby(["COUNTY", "year"], as_index=False)["votes"].sum().rename(columns={"votes": "total_votes"}))
```

County Vote Distribution 

```python
county_party_all = county_party_all.merge(county_totals, on=["COUNTY", "year"], how="left")
county_party_all
```

```python
county_party_all["vote_share"] = (county_party_all["votes"] / county_party_all["total_votes"])
county_party_all
```

```python
county_party_sorted = county_party_all.sort_values(["COUNTY", "year", "votes"],ascending=[True, True, False])
```

```python
top_two = (county_party_sorted.groupby(["COUNTY", "year"]).head(2))
top_two
```

Win Margin

```python
margin = (top_two.groupby(["COUNTY", "year"]).apply(lambda x: x.iloc[0]["votes"] - x.iloc[1]["votes"]).reset_index(name="margin_votes"))
margin
```

```python
winners = (top_two.groupby(["COUNTY", "year"]).first().reset_index()[["COUNTY", "year", "party"]].rename(columns={"party": "winning_party"}))
winners
```

```python
county_results = (winners.merge(margin, on=["COUNTY", "year"]).merge(county_totals, on=["COUNTY", "year"]))
county_results
```
